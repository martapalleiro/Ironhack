## Map, filter & reduce

### Map

```python
import math

def area(r):
    return math.pi * (r**2)
```

```python
radii = [2, 5, 7.1, 0.3, 10]
```

#### Example 1

```python
# Method 1: direct method

areas = []
for r in radii:
    a = area(r)
    areas.append(a)
areas
```

```python
# Method 2: using the 'map' function

map(area, radii) # map(function, list)
```

```python
# the output is not a list, but a map object
list(map(area, radii))
```

#### Example 2

```python
n = [4,3,2,1]
```

```python
# traditional method
def square(lst1):
    lst2 = []
    for num in lst1:
        lst2.append(num ** 2)
    return lst2
```

```python
print(square(n))
```

```python
list(map(lambda x: x**2, n))
```

```python
# with a list comprehension
[x**2 for x in n]
```

### Filter

```python
import statistics

data = [1.5, 9.2, 3.4, 5.1, -9,4]

avg = statistics.mean(data)
avg
```

```python
filter(lambda x: x > avg, data)
```

```python
list(filter(lambda x: x > avg, data))
```

```python
def above_avg(x):
    return x > avg
```

```python
above_avg(1)
```

```python
above_avg(5)
```

```python
list(filter(above_avg, data))
```

```python
# with a list comprehension

[i for i in data if i > avg]
```

### Reduce

```python
from functools import reduce

names = ["Guillem", "Ali", "Anna", "Rapha"]

concatenator = lambda x, y: x + y

reduce(concatenator, names)
```

```python
numbers = [1,2,3,4]
```

```python
reduce(concatenator, numbers)
```

```python
reduce(lambda x,y: x*y, numbers)
```
