```python
# import pandas and numpy
import pandas as pd
import numpy as np
```

```python
"""pd.Series(list)"""
# Convert list into Series (!!! write Series capitalized) --> pd.Series(list)
# data type: floating

float_list = [1.0, 3.0, 8.3, 9.88]
my_float_series = pd.Series(float_list)
print(my_float_series)
```

```python
"""Slicing and indexing"""
# Works as always
# data type: object (type is decided based on the most restrictive value: a string can only be interpreted 
           # as a string, so now everything is a string)

my_float_series[1]="hello"
print(my_float_series)
```

```python
"""pd.DataFrame(dict)"""
# To create dataframe from lists, convert first the lists to dicionnaires, where keys are the titles of the variables

my_ints = [1,2,3,4]
float_list = [1.0, 3.0, 8.3, 9.88]
my_strings = ["hey", "ho", "lets", "go"]

my_dict = {"ints": my_ints,
          "floats": float_list,
          "strings": my_strings}

my_df = pd.DataFrame(my_dict)
print(my_df)
```
